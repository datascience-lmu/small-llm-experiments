```python
import polars as pl
import matplotlib.pyplot as plt
```

```python
l1 = pl.read_csv("list1.csv") # List sorting, no chain of though
l2 = pl.read_csv("list2.csv") # List sorting, chain of though
l3 = pl.read_csv("list3.csv") # List sorting, no chain of though
l4 = pl.read_csv("list4.csv") # List sorting, chain of though
l5 = pl.read_csv("list5.csv") # List sorting increasing, no chain of though
```

```python
# List sorting, no chain of though
f = l1.filter(pl.col("number of digits") == 1)
plt.plot(f["number of words"], f["success count"] / (f["success count"] + f["fail count"]), label='1 digit')
f = l1.filter(pl.col("number of digits") == 2)
plt.plot(f["number of words"], f["success count"] / (f["success count"] + f["fail count"]), label='2 digits')
f = l1.filter(pl.col("number of digits") == 3)
plt.plot(f["number of words"], f["success count"] / (f["success count"] + f["fail count"]), label='3 digits')
f = l1.filter(pl.col("number of digits") == 4)
plt.plot(f["number of words"], f["success count"] / (f["success count"] + f["fail count"]), label='4 digits')
plt.xlabel("Number of Elements")
plt.ylabel("Success Probability")
plt.legend()
plt.show()
```
